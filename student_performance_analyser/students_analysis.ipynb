```python
read["TOTAL_MARKS"]= read["MATHEMATICS"]+read["SCIENCE"]+read["APTITUDE"]
```

```python
read["AVERAGE"]=read["TOTAL_MARKS"]/3
read["GRADE"]=np.where(read["AVERAGE"]>=80,"A",np.where(read["AVERAGE"]>=70,"B",np.where(read["AVERAGE"]>=50,"C","F")))
''' def assign_grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "Fail"

df["GRADE"] = df["AVERAGE"].apply(assign_grade)'''
```

```python
read.sort_values(by='AVERAGE',ascending=False).iloc[0]["STUDENT_NAME"]#who performs the best in class?
```

```python
math_avg=read['MATHEMATICS'].mean()# which subject is easier nd which subject is harder?
math_avg
```

```python
science_avg=read["SCIENCE"].mean()# math is hard. science is the easiest 
science_avg
```

```python
aptitude_avg=read["APTITUDE"].mean()
aptitude_avg
```

```python
# lets calculate the standard deviation for each subject and check its consistency with students
```

```python
a = np.std(read["MATHEMATICS"])
a
```

```python
b =np.std(read["SCIENCE"])
b
```

```python
c = np.std(read["APTITUDE"])
c
```

```python
import matplotlib.pyplot as plt
```

```python
subjects = ["Mathematics", "Science", "Aptitude"]
std_values = [a,b,c]

plt.bar(subjects, std_values)
plt.xlabel("Subjects")
plt.ylabel("Standard Deviation")
plt.title("Standard Deviation of Marks by Subject")
plt.savefig("plots/std_deviation_subjects.png")

```

```python
sub_avg=[math_avg,science_avg,aptitude_avg]
plt.bar(subjects, sub_avg)
plt.xlabel("Subjects")
plt.ylabel("Subjects Average!")
plt.title("Subjects Average of Subject")
plt.savefig("plots/subject_average.png")
plt.show()
```

```python
grade_counts = read["GRADE"].value_counts()

plt.pie(
    grade_counts.values,
    labels=grade_counts.index,
    autopct="%1.1f%%"
)

plt.title("Grade Distribution of Students")
plt.savefig("plots/grade_distribution.png")
plt.show()
```
